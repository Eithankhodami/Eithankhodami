```python
#imagine we want to have a new column which includes the following data:  new_column = SepalLengthCm / SepalWidthCm and will be added to the last column:
#for this act we use assign() command:

dataset.assign(SepalNewData = dataset['SepalLengthCm'] / dataset['SepalWidthCm'])
```

```python
#remember that pandas data is DataFrame format
```

```python
#the action we did upper also can be done by lambda
#for more practice we create a new colum:
dataset.assign(sqrt_form = lambda x: (x['SepalWidthCm']/x['PetalWidthCm'] **2))
#this is to remind that assign can handle a function
#NOTE: assign does not change the original data, it returns a copy of data
```

```python
#if you want  to check some data for example which data is bigger than 5, which is smaller than 3 or so on, we can use q query format:
dataset.query('SepalLengthCm > 5')
#this is more like a boolean data and while it is true, it shows the cell.
```

```python
#we have option of combinig query with a function
```

```python
dfa = pd.DataFrame(
{
    'A': [1, 2, 3],
    'B' : [4, 5, 6]

}
)
print(dfa)
```

```python
#now if I want to add a new column 'c' and then ask pandas to add 'D' which is 'C' and 'A'
#I can do this just in one lambda function:
dfa.assign(C = lambda x : x['A'] + x["B"], D = lambda x : x['A'] + x['C'])
```
